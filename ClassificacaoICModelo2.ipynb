```python
import pymysql.cursors
```

```python
#conectando com o banco de dados 
conn = pymysql.connect( user='root', password='', db='ic_comentarios' )

#inserindo
base = []
try: 
    with conn.cursor() as cursor: 
        query = "SELECT `comentario`, `emocao` FROM `tb_comentarios`"
        cursor.execute(query)
        result = cursor.fetchall()
        for linha in result:
            print(linha)
            base.append(linha)
except:
    print('erro')

finally:
    conn.close()
```

```python
import nltk
```

```python

stopwords = nltk.corpus.stopwords.words('portuguese')

stemmer = nltk.stem.RSLPStemmer()

#extração de radical + remoção de stopwords

def stopRadical(base):
        frases = []
        for(frase, emocao) in base:
                comStemmer = [str(stemmer.stem(p)) for p in frase.split() if p not in stopwords] 
                frases.append((comStemmer, emocao))
        return frases

newBase = stopRadical(base)
print(newBase)
```

```python
#listando palavras

def busca(base):
        palavras = []
        for(frase, emocao) in base:
                palavras.extend(frase)
        return palavras

palavras = busca(newBase)
print(palavras)
```

```python
#retornando palavras únicas

#obtendo a frequencia das palavras

def frequencia(palavras):
        palavra = nltk.FreqDist(palavras)
        return palavra

frequencia = frequencia(palavras)
print(frequencia.most_common(50))
```

```python
#retornando unicas

def unicas(frequencia):
        freq = frequencia.keys()
        return freq

palavras_unicas = unicas(frequencia)

print(palavras_unicas)
```

```python
#extração das palavras. Tenho ou não tenho?

def extra_palavras(frase):
        doc = set(frase) #estrutura sem repetição de dados
        caracteristicas = {}
        for palavra in palavras_unicas:
                caracteristicas['%s' % palavra] = (palavra in doc)
        return caracteristicas

base_completa = nltk.classify.apply_features(extra_palavras, newBase)
print(base_completa[0])
```

```python
#usando Naive Bayes 

#criando a tabela de probabilidade 

classificador = nltk.NaiveBayesClassifier.train(base_completa)
print(classificador.labels()) #traz as classes, ou seja, as ameções.
#traz as características importantes. 
#como leio? quando dia é True, tenho 2.3 de probabilidade de ser uma frase de alegria e 1.0 de ser medo
print(classificador.show_most_informative_features()) 
```

```python
#entrando com uma frase sem a definição da emoção

teste = "eu fui espancado porque sou travesti"
```

```python
palavraTeste = []
for palavrasTeste in teste.split():
        testeStm = [p for p in palavrasTeste.split()]
        palavraTeste.append(str(stemmer.stem(testeStm[0])))
```

```python
novaTeste = extra_palavras(palavraTeste)

print(classificador.classify(novaTeste))
```

```python
#classificando com probabilidade
dist = classificador.prob_classify(novaTeste)
for classe in dist.samples():
        print("%s %f" % (classe, dist.prob(classe)*100))
```
